## Clean Data


```python
# Drop columns
columns_to_drop = ["id", "name", "host_name", "last_review"]
df.drop(columns=columns_to_drop, inplace=True)

df.columns
```

```python
df.dropna().info()
```

```python
df.dropna(subset=df.columns.difference(["reviews_per_month"])).info()
```

```python
# Fill missing values by drawing from a random sample of non-null values of the same column.
# NOTE: Always be careful when filling missing values, as it can introduce bias into your dataset.
#   In this case, we are filling missing values with random values from the same column,
#   in order to preserve the distribution of the data. However, this will not always be the case.
def fillna_with_random(series):
    mask = series.isnull()
    non_null_values = series[~mask].values
    random_values = np.random.choice(
        non_null_values, size=mask.sum(), replace=True
    )
    series[mask] = random_values
    return series


df = df.apply(fillna_with_random, axis=0)

# Drop duplicate rows
df.drop_duplicates(inplace=True)

df.info()
```

```python
display(df["neighbourhood_group"].unique())

# Fix typos in the "neighbourhood_group" column
df["neighbourhood_group"].replace("manhatten", "Manhattan", inplace=True)
df["neighbourhood_group"].replace("Brooklin", "Brooklyn", inplace=True)

df["neighbourhood_group"].unique()
```

```python
# Remove outliers from the "price" column
display(df["price"].describe())

df = df[(df["price"] > 10) & (df["price"] < 250)]

df["price"].describe()
```

```python
# Fix datatype of "number_of_reviews" column.
display(df["number_of_reviews"].dtype)
str_mask = df["number_of_reviews"].apply(lambda x: isinstance(x, str))
display(df["number_of_reviews"][str_mask])

df["number_of_reviews"] = df["number_of_reviews"].str.replace(" reviews", "")
df["number_of_reviews"] = df["number_of_reviews"].astype(
    float, errors="ignore"
)

df["number_of_reviews"].dtype
```

## Distribution of Prices


```python
plt.hist(df["price"], bins=50, range=(0, 1000))
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

## Top 10 Neighborhoods With Highest Number of Listings


```python
top_neighborhoods = df["neighbourhood"].value_counts().head(10)
top_neighborhoods.plot(kind="bar")
plt.xlabel("Neighbourhood")
plt.ylabel("Number of Listings")
plt.title("Top 10 Neighbourhoods by Listings")
plt.show()
```

## Relationship Between Price and Availability


```python
plt.scatter(df["price"], df["availability_365"], alpha=0.1)
plt.xlabel("Price")
plt.ylabel("Availability")
plt.title("Price vs Availability")
plt.show()


print(
    "Correlation between price and availability: ",
    df.corr(numeric_only=True).loc["price", "availability_365"],
)
```

# Average Price by Neighborhood and Room Type


```python
df.groupby(["neighbourhood", "room_type"]).mean()["price"]
```

## Availability of Listings by Neighborhood


```python
availability_pivot = df.pivot_table(
    index="neighbourhood", columns="room_type", values="availability_365"
).fillna(0)
sns.clustermap(availability_pivot)  # Requires scipy to be installed
plt.title("Availability Heatmap by Neighbourhood and Room Type")
plt.show()
```

## Top 10 Hosts With the Most Listings


```python
top_hosts = df["host_id"].value_counts().head(10)
top_hosts.plot(kind="bar")
plt.xlabel("Host ID")
plt.ylabel("Number of Listings")
plt.title("Top 10 Hosts by Listings")
plt.show()
```

## Relationship Between the Number of Reviews and Price


```python
plt.scatter(df["price"], df["number_of_reviews"], alpha=0.1)
plt.xlabel("Price")
plt.ylabel("Number of Reviews")
plt.title("Price vs Number of Reviews")
plt.show()

print(
    "Correlation between price and number of reviews: ",
    df.corr(numeric_only=True).loc["price", "number_of_reviews"],
)
```
